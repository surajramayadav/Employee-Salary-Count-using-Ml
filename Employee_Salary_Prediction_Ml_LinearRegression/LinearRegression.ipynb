```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('salary.csv')
```

```python
df.head()
```

```python
plt.scatter(df["Exprience"],df["salary"])
plt.xlabel('Exprience')
plt.xlabel('Salary')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df[["Exprience"]],df['salary'],test_size=0.3)
```

```python
reg=LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
ytest
```

```python
reg.predict(xtest)
```
